# 1. Getting index data

some famous us-market indices are :
* S&P 500 : ticker=^GSPC
* Dow Jones : ticker=^DJI

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import time
```

```python
tickers = ["^GSPC", "^DJI"]
indexes = yf.download(tickers, period='5y')
indexes
```

as for how the above data can be grouped according to ticker and make it more  usable : refet to 1.GettingStockDataFromYfinance. subarticle:8 group by ticker

```python
# getting only closing data
tickers = ["^GSPC", "^DJI"]
indexes = yf.download(tickers, period='5y').Close
indexes
```

```python
indexes.iloc[0].mul(100)
```

```python
normalizedData = indexes.div(indexes.iloc[0]).mul(100)
normalizedData
```

```python
normalizedData.plot()
plt.show()
```

<b><u> and this process of normalization is universal i.e 2 or more assets may not be homogeneous i.e one can be equity , another can index, while another can be equity . thus this normaization is a pretty powerful tool that helps us to normalize and compare. </u></b>

# 6. getting real-time data

```python
df = yf.download("EURJPY=X", period='1d', interval='1m')
```

```python
# to get display the last few records 
df.tail()
```

```python
#to get the last index
print(df.index[-1])
```

```python
#to print the last index and close price
print(df.index[-1], df.iloc[-1, 3]) 
#i.e. in iloc() we are passing the rowindex and the columnindex
```

now combining the above techniques we can get the latest data streamed as shown below. 
* here i have polled only 5 times using a counter
* however to run this live streamer we can have an infinite while loop using while True
* if we want the timer to run only during market hours when we have to write the while loop accordingly

```python
counter = 5
while counter>0:
    time.sleep(1) # sleeping for 5 seconds
    data = yf.download("EURJPY=X", period='1d', interval='1m')
    print(data.index[-1], data.iloc[1,3])
    counter= counter-1
```

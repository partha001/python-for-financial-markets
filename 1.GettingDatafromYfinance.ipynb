# Importing data from yahoo finance

# 1. Getting historical-data [i.e.  Open/high/low/close] using yfinance 

```python
import pandas as pd
import yfinance as yf
```

```python
# go to yahoo finance and find the symbol/ticker for the instrument
ticker = "TSLA"
yf.download(ticker)
```

do shift + tab inside the yf.download() method to explore all the parameters.

# 2. Getting intra-day data

## getting intraday data for 1h interval

```python
#for getting the intra-day data we will make use of the interval parameter of the yf.downloadl
#as shown below

#however it is to be kept in mind that when using the interval the duration can not extends beyond 
#last 60 days
ticker="ITC.NS"
#getting 1hr data for last 1hr
yf.download(ticker, period ='1mo', interval='1h')
```

## getting intra-day data for 5min interval

```python
# similarly getting 30min data for lat one month
yf.download(ticker, period='1mo', interval = '5m')
```

note that when we fetched data for 30min interval we got 161 records. however when we fetched 5min interval data we received 1725 records for the same duration i.e. 1month

## getting intra-data for 1min interval

```python
# getting 1min data for last 5days
yf.download(ticker, period='5d', interval='1m')
```

* however with yahoo finance we have to keep in mind that for a particualr stock if there is no activity then we will not get the data though its a valid ticker and it exists in the market for that min [when there is no transaction]
* it is also to be noted that the time that we receive in datetime is in UTC , and gives us the record from 9:15AM IST to 3:29PM IST

```python
## exporting the data to csv
itc = yf.download(ticker, period='5d', interval='1m')
itc.to_csv("./data-export/exported-itc.csv")
```

# 3. Getting Pre-market and post-market data

```python
## US market timings

# pre-market trading time
# 4:00AM EST to 9:29 am EST

# regular market trading time
# 9:30 AM EST to 4:00 PM EST

# post-market trading time
# 4.01PM EST to 8:00PM EST
```

```python
# thus for getting the pre and post-market data we have to pass prepost=True . the default value of this parameter is False
# and hence by default the download() gives us only regular-market trading time data
ticker = "GOOG"
yf.download(ticker, period='5d', interval='1m', prepost=True)
```

```python
# however this is to be noted that this does not work for Indian-market and it gives us only regular-market-hour data
# for indian tickers even if we pass this parameter as True
ticker = "RELIANCE.NS"
yf.download(ticker, period='5d', interval='1m', prepost=True)
```

# 5.  getting dividends and stock-splits

## getting dividend and stock-split information

```python
# since dividends and stock-splits are also called corporate=actions so 
#we can fetch them as shown below
ticker = "RELIANCE.NS"
reliance = yf.Ticker(ticker)
reliance.actions
```

## getting only dividend data

```python
ticker = "RELIANCE.NS"
reliance = yf.Ticker(ticker)
reliance.dividends
```

## getting only stock-split data

```python
ticker = "RELIANCE.NS"
reliance = yf.Ticker(ticker)
reliance.splits
```

## getting dividend and stock-split info as part of historical data

```python
# the dividends and stock splits are also sometime referred to as corporate actions
# and in case if we want to get it has part of historical data then we have to pass 
# action=True  to the download method. the value of this parameter is false

ticker = 'GOOG'
yf.download(ticker, period='10y', actions=True)
```

```python
#lets verify the same on some indian ticker
ticker = 'ITC.NS'
yf.download(ticker, period='10y', actions=True)
```

so this works equally well for indian symbols

### filtering dividends and stock-splits from historical data

```python
#we can also verify this applying filter on the dataframe
ticker = 'ITC.NS'
itc = yf.download(ticker, period='10y', actions=True)
itc.columns
```

```python
# filtering records for dividends
itc[itc[('Dividends', 'ITC.NS')]>0].head(5)
```

```python
# filtering to see only stock-split records
itc[itc[('Stock Splits', 'ITC.NS')]>0].head(5)
```

```python
#combining condition
itc[(itc[('Dividends', 'ITC.NS')]>0) | (itc[('Stock Splits', 'ITC.NS')]>0)].head(5)
```
